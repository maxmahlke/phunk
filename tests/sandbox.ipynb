# SOCCA

```python
data = pd.read_csv('data/Eugenia_ztf.csv')
cols = ['Date','cmagpsf']

sol =  pd.read_csv('data/Eugenia_socca.csv')
```

```python
data.columns[ data.columns.str.contains('f') ]
```

```python
pc = phunk.PhaseCurve(
    target="Eugenia",
    epoch=data["Date"].values - 2400000.5,
    phase=data["Phase"].values,
    mag=data["cmagpsf"].values,
    band=data["cfid"].values,
)
```

```python
pc.get_ephems()
```

```python
pc.fit(models=['HG1G2'])
```

```python
pc.HG1G2.H1
```

```python
pc.fit(models=['SOCCA'], pinit='toto')
```

```python
pc.SOCCA.H = sol.loc[0, "H_1"]
pc.SOCCA.G1 = sol.loc[0, "G1_1"]
pc.SOCCA.G2 = sol.loc[0, "G2_1"]
pc.SOCCA.period = sol.loc[0, "period"]
pc.SOCCA.alpha = sol.loc[0, "alpha0"]
pc.SOCCA.delta = sol.loc[0, "delta0"]
pc.SOCCA.a_b = sol.loc[0, "a_b"]
pc.SOCCA.a_c = sol.loc[0, "a_c"]
pc.SOCCA.t0 = 2459580.5 # Hard coded in fink utils
pc.SOCCA.W0 = sol.loc[0, "phi0"]
```

```python
pc.fit()
```
