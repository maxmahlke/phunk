```python
pc = phunk.PhaseCurve(
    target="Eugenia",
    epoch=data["Date"].values,
    phase=data["Phase"].values,
    mag=data["cmred"].values,
    band=data["cfid"].values,
)
```

```python
pc.fit(models=["SOCCA"], p0=p0, remap=True)
```
